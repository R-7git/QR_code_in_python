```python
# Post Request


'''

HTTP POST Request:
1) POST is one of the HTTP methods (others include GET, PUT, DELETE, etc.).
2) It's used to send data to a server to create/update a resource.
3) In web applications, POST requests are commonly used when submitting forms or
4) when sending data that modifies the state of the server/database.

When to Use:
1) Use requests.post when you need to send data to a server in a way that's not
   visible in the URL (unlike GET requests where parameters are in the URL).

2) Typical use cases include submitting form data, uploading files, or any scenario
   where you need to send structured data to a server.

Example Use Case:
) If you were building a web application with a form where users submit their 
  name and email, you would use a POST request to send this data to your server 
  for processing and storing in a database.

'''
```

```python
# Note : 'With' class if usd in files , we dont need to use f.close() method 

url = "https://assets.mixkit.co/active_storage/sfx/213/213.wav"

# url= "https://www.win-rar.com/fileadmin/winrar-versions/winrar/winrar-x64-621.exe"

r = requests.get(url, stream=True)

totalExpectedBytes = int(r.headers["Content-Length" ])

#print(totalExpectedBytes)

bytesReceived=0

with open("winrar.wav",'wb') as f:
    for chunk in r.iter_content(chunk_size=128):
        print(f" {bytesReceived} received out of total {totalExpectedBytes}")
        f.write(chunk)
        bytesReceived +=128
```

```python

# we can see the bar but try to download the larger file ,
# cos the bar stays longer n clearly visible 

url = "https://assets.mixkit.co/active_storage/sfx/213/213.wav"

r = requests. get (url,stream=True)
totalExpectedBytes = int(r.headers["Content-Length" ])
# print(totalExpectedBytes)
bytesReceived = 0
progress_bar = tqdm(total=totalExpectedBytes, unit='iB', unit_scale=True)
with open ("winrar.exe", 'wb') as f:
    for chunk in r.iter_content(chunk_size=128):
    # print(f"-{bytesReceived} received out of total {totalExpectedBytes}")
        progress_bar.update(128)
        f.write(chunk)
        bytesReceived += 128
progress_bar.close() 

```
